## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
import re

def convert_case(text: str, force_format: str = None) -> str:
    if not text:
        return text
        
    is_snake = '_' in text or text.islower()
    is_pascal = not is_snake and text[0].isupper()
    
    if force_format == 'snake':
        if is_snake: return text
        return re.sub(r'(?<!^)(?=[A-Z])', '_', text).lower()
    elif force_format == 'pascal':
        if is_pascal: return text
        return ''.join(word.capitalize() for word in text.split('_'))
        
    if is_snake:
        return ''.join(word.capitalize() for word in text.split('_'))
    else:
        return re.sub(r'(?<!^)(?=[A-Z])', '_', text).lower()


print(convert_case("otus_course")) 

print(convert_case("PythonIsTheBest")) 
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
from datetime import datetime


def is_valid_date(date_str: str) -> bool:
    try:
        datetime.strptime(date_str, "%d.%m.%Y")
        return True
    except ValueError:
        return False


# Примеры из задания — результат появится под ячейкой
# is_valid_date("29.02.2000")
is_valid_date("29.02.2001")
# is_valid_date("31.04.1962")
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
import math


def is_prime(n: int) -> bool:

    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    for i in range(3, int(n ** 0.5) + 1, 2):
        if n % i == 0:
            return False
    return True



is_prime(17)
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
def normalize_name(value):
    if not value.isalpha():
        return None
    return value.capitalize()


def read_age():
    while True:
        age_input = input("Возраст: ").strip()
        if not age_input:
            return None

        if not age_input.isdigit():
            print("Ошибка: возраст должен быть числом от 18 до 60.")
            continue

        age = int(age_input)
        if 18 <= age <= 60:
            return age

        print("Ошибка: возраст должен быть числом от 18 до 60.")


def read_user_id(existing_ids):
    while True:
        user_id_input = input("ID: ").strip()
        if not user_id_input:
            return None

        if not user_id_input.isdigit():
            print("Ошибка: ID должен быть целым числом.")
            continue

        user_id = f"{int(user_id_input):08d}"
        if user_id in existing_ids:
            print("Ошибка: такой ID уже существует. Введите другой.")
            continue

        return user_id


def collect_users():
    users = {}

    while True:
        name_input = input("Имя: ").strip()
        if not name_input:
            break

        name = normalize_name(name_input)
        if name is None:
            print("Ошибка: имя должно состоять только из букв.")
            continue

        surname_input = input("Фамилия: ").strip()
        if not surname_input:
            break

        surname = normalize_name(surname_input)
        if surname is None:
            print("Ошибка: фамилия должна состоять только из букв.")
            continue

        age = read_age()
        if age is None:
            break

        user_id = read_user_id(set(users.keys()))
        if user_id is None:
            break

        users[user_id] = (name, surname, age)

    return users


def print_users_table(users):
    if not users:
        print("Пользователи не добавлены.")
        return

    headers = ("ID", "Имя", "Фамилия", "Возраст")
    rows = [
        (user_id, name, surname, str(age))
        for user_id, (name, surname, age) in sorted(users.items())
    ]

    widths = [
        max(len(headers[i]), *(len(row[i]) for row in rows))
        for i in range(len(headers))
    ]

    header_line = " | ".join(headers[i].ljust(widths[i]) for i in range(len(headers)))
    separator = "-+-".join("-" * widths[i] for i in range(len(headers)))

    print(header_line)
    print(separator)
    for row in rows:
        print(" | ".join(row[i].ljust(widths[i]) for i in range(len(headers))))


print("Ввод пользователей. Пустое поле завершает ввод.")
users = collect_users()
print("\nИтоговая таблица:")
print_users_table(users)
```
